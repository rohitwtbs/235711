```python
import pandas as pd
df = pd.read_csv("e_read.csv")
print(df)
```

```python
df.columns = ["id", "sensor_name", "status", "time"]
```

```python
print(df)
```

```python
print(df.groupby('sensor_name'))
```

```python
gp_df = df.groupby('sensor_name')
print(gp_df.first())
```

```python
for sensor , sensor_df in gp_df:
    print(sensor)
    print(sensor_df)
```

```python
print(gp_df.get_group('Sensor3ee283'))
```

```python
print(Sensor3ee283_df)
print(Sensor3ee283_df.count())
```
